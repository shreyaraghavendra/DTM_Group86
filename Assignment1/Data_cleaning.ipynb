## Outlier Detection and Removal
Paper: https://www.sciencedirect.com/science/article/pii/S1574013720304068

```python
def plot_outliers_scatter(df, nrows, ncols, categories): 
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(12, 12))  
    for i in range(nrows):
        for j in range(ncols):
            category_index = i * 3 + j  # Calculate the index for accessing categories list
            category_name = categories[category_index]  # Get the category name
            
            axes[i, j].scatter(df['id'], df[category_name])
            axes[i, j].set_xlabel('id', fontweight='bold')
            axes[i, j].set_ylabel(category_name, fontweight='bold')

    plt.tight_layout()
    plt.show()

categories = ['appCat.communication', 'appCat.entertainment', 'appCat.social', 'appCat.finance',
                'appCat.game', 'appCat.office', 'appCat.social', 'appCat.travel', 'appCat.weather',
                'appCat.builtin', 'appCat.utilities', 'appCat.other', 'appCat.unknown', 'screen','mood']

plot_outliers_scatter(pivot, 5, 3, categories)
```

```python
def remove_outliers_iqr(df, columns, multiplier=1.5):
    """
    Removes outliers from a dataframe based on the IQR method.
    
    Parameters:
    - df: pandas DataFrame.
    - columns: list of columns to check for outliers.
    - multiplier: the multiplier for the IQR to adjust strictness.
    
    Returns:
    - DataFrame with outliers set to None.
    """
    df_outliers_removed = df.copy()
    bounds = {}
    
    for column in columns:
        Q1 = df_outliers_removed[column].quantile(0.25)
        Q3 = df_outliers_removed[column].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - (multiplier * IQR)
        upper_bound = Q3 + (multiplier * IQR)
        
        df_outliers_removed[column] = df[column].apply(lambda x: x if x >= lower_bound and x <= upper_bound else None)
        bounds[column] = (lower_bound, upper_bound)
    
    return df_outliers_removed, bounds

columns_to_check = ['appCat.communication', 'appCat.entertainment', 'appCat.social', 'appCat.finance',
                'appCat.game', 'appCat.office', 'appCat.social', 'appCat.travel', 'appCat.weather',
                'appCat.builtin', 'appCat.utilities', 'appCat.other', 'appCat.unknown', 'screen']
```

```python
df_cleaned, outlier_bounds = remove_outliers_iqr(pivot, columns_to_check, 70)
plot_outliers_scatter(df_cleaned, 5, 3, columns_to_check)
```

## Average all recordings of all attributes taken in one day

```python
df_cleaned['time'] = pd.to_datetime(df_cleaned['time']) #convert string to datetime
averaged_df = df_cleaned.groupby(['id', df_cleaned['time'].dt.date]).mean() # average all values taken on the same day
averaged_df = averaged_df.drop('time', axis=1)
averaged_df
```

```python
df_reset = averaged_df.reset_index()
heatmap_df = df_reset.pivot(index='time', columns='id', values='mood')

# Set the background color to white for NaN values
sns.set_style("white")

# Create the heatmap
plt.figure(figsize=(10, 8))  # Adjust the size as needed
ax = sns.heatmap(heatmap_df, cmap="viridis", linewidths=.5, linecolor='gray', cbar_kws={'label': 'Mood'})

# Set the colors for NaN values to white
ax.set_facecolor('white')

plt.title('Heatmap of Mood by Day and ID')
plt.xlabel('ID')
plt.ylabel('Day')
plt.show()
```

```python
# Set seaborn style for the plots
sns.set(style="white")

# List all variable names except 'id' and 'time' which are not to be plotted
variables = [col for col in df_reset.columns if col not in ['id', 'time']]

# Prepare for plotting
num_variables = len(variables)
num_rows = -(-num_variables // 5)  # Ceiling division to calculate the number of rows needed
fig, axs = plt.subplots(num_rows, 5, figsize=(20, num_rows * 4))  # Adjust figsize as needed

# Flatten the axs array for easy iteration
axs = axs.flatten()

for i, var in enumerate(variables):
    # Pivot the table for the current variable
    heatmap_df = df_reset.pivot(index='time', columns='id', values=var)
    
    # Plot the heatmap
    sns.heatmap(heatmap_df, ax=axs[i], cmap="viridis", cbar_kws={'label': var}, linewidths=.5, linecolor='gray')
    axs[i].set_title(var)
    axs[i].set_facecolor('white')  # Set the background color for NaN values to white

# Hide any unused subplots that might exist if the number of variables is not a multiple of 5
for j in range(i + 1, num_rows * 5):
    axs[j].axis('off')

plt.tight_layout()
plt.show()
```

# Imputation

### Simple Imputation

```python
df = pd.read_csv('data_cleaned.csv')
df_pivot = df.pivot(index='time', columns='id', values='mood')
df_mood_simple = df_pivot.ffill(limit=5)

# plot heatmap
heatmap_df = df_mood_simple
sns.set_style("white")
plt.figure(figsize=(10, 8))
ax = sns.heatmap(heatmap_df, cmap="viridis", linewidths=.5, linecolor='gray', cbar_kws={'label': 'Mood'})

# Set the colors for NaN values to white
ax.set_facecolor('white')

plt.title('Heatmap Mood, Simple Imputation')
plt.xlabel('ID')
plt.ylabel('Day')
plt.show()
```

### Rolling Window Imputation

```python
window_size = 5
df_mood_rolling = df_pivot.fillna(df_pivot.rolling(window=window_size, min_periods=1).mean())

heatmap_df = df_mood_rolling
sns.set_style("white")

# Create the heatmap
plt.figure(figsize=(10, 8))  # Adjust the size as needed
ax = sns.heatmap(heatmap_df, cmap="viridis", linewidths=.5, linecolor='gray', cbar_kws={'label': 'Mood'})

# Set the colors for NaN values to white
ax.set_facecolor('white')

plt.title('Heatmap Mood, Rolling Window (5 days)')
plt.xlabel('ID')
plt.ylabel('Day')
plt.show()
```

```python
df_sorted = df.sort_values(by=['time', 'id'])
features = list(df_sorted.columns)
features.remove('id')
features.remove('time')

df_simple = df_sorted.copy()
df_rolling_window = df_sorted.copy()

# Simple imputation
for feature in features:
    df_simple[feature] = df_sorted[feature].fillna(method='ffill', limit=5)

# Rolling window
window_size = 5
for feature in features:
    df_rolling_window[feature] = df_sorted[feature].fillna(df_sorted[feature].rolling(window=window_size, min_periods=1).mean())

df_simple.to_csv('df_simple_imputation.csv', index=False)
df_rolling_window.to_csv('df_rolling_window_imputation.csv', index=False)
```
